```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_data = pd.merge(mouse_metadata,study_results, how='right',left_on = "Mouse ID", right_on = "Mouse ID")

# Display the data table for preview
display(combined_data.head())
```

## Summary Statistics

```python
clean_df
df_mean=clean_df.groupby("Drug Regimen").mean("numerical")["Tumor Volume (mm3)"]
display(df_mean)
```

```python
df_median=clean_df.groupby("Drug Regimen").median("numerical")["Tumor Volume (mm3)"]
display(df_median)
```

```python
df_var=clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
display(df_var)
```

```python
df_std=clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
display(df_std)
```

```python
df_sem=clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()
display(df_sem)
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
pd_concat = pd.concat([df_mean, df_median, df_var, df_std, df_sem],axis=1)
pd_concat.columns = ["Mean Tumor Volume", "Median Tumor Volume", "Tumor Volume Variance", "Tumor Volume Std. Dev.", "Tumor Volume Std. Err."]
pd_concat
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
# Using the aggregation method, produce the same summary statistics in a single line
clean_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(["mean","median","var","std","sem"])
```

## Quartiles, Outliers and Boxplots

```python
clean_df
```

```python
# Start by getting the last (greatest) timepoint for each mouse
max_time = clean_df.groupby("Mouse ID").max()["Timepoint"].to_frame()
max_time
```

```python
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
max_merge = pd.merge(max_time,clean_df,left_on=[max_time.index, "Timepoint"],right_on=["Mouse ID","Timepoint"])
max_merge
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
# Create empty list to fill with tumor vol data (for plotting)
tumor=[]
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in treatments:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    temp = max_merge[max_merge["Drug Regimen"]== drug]["Tumor Volume (mm3)"]
    # add subset
    tumor.append(temp)
     # Determine outliers using upper and lower bounds
    q1=temp.quantile(0.25)
    q3=temp.quantile(0.75)
    IQR=q3-q1
    upperbound=q3+1.5*IQR
    lowerbound=q1-1.5*IQR
    outlier=(temp > upperbound)|(temp < lowerbound)
    print(drug,"'s potential outliers:",temp [outlier])
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
plt.boxplot(tumor)
plt.ylabel("Final Tumor Volume(mm3)")
plt.xticks([1,2,3,4],treatments)
plt.show()
```

```python
weightvs_tumor = clean_df[clean_df["Drug Regimen"]=="Capomulin"].groupby("Mouse ID").mean("numeric")
weightvs_tumor
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
weightvs_tumor.plot(x= "Weight (g)",y="Tumor Volume (mm3)", kind="scatter")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
slope, intercept, rvalue, pvalue, stderr = st.linregress(weightvs_tumor["Weight (g)"],weightvs_tumor["Tumor Volume (mm3)"])
print("The correlation between mouse weight and the average tumor volume is ",round(rvalue,2))
plt.scatter(weightvs_tumor["Weight (g)"],weightvs_tumor["Tumor Volume (mm3)"])
x=weightvs_tumor["Weight (g)"]
y=slope*x+intercept
plt.plot(x,y, color="red")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```
